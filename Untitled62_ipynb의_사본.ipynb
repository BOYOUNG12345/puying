```python
#분리 개수 확인 코드
from collections import Counter

with open('train.csv', encoding='utf-8') as f:
  lines = f.readlines()
print(len(lines))

len_list=[]
for _ in range(len(lines)):
  print(str(_)+":"+str(len(lines[_].split(','))))
  len_list.append(len(lines[_].split(',')))

Counter(len_list)
```

```python
#해결방법 -> csv패키지를 import하고, csv파일을 읽은 pandas DataFrame으로 만들어버리는 방법

import csv #내장된 csv모듈을 import
import pandas as pd
f= open('test.csv', encoding='utf8') #파일 열기
reader= csv.reader(f) #iterator 타입인 reader객체를 리턴-> for루프를 돌며 한 라인씩 가져올 수 있음// 리턴되는 각 라인은 컬럼들을 나열한 list타입
csv_list=[]
for i in reader:
  csv_list.append(i)
f.close()
test_df= pd.DataFrame(csv_list)
test_df.info()
```

```python
#train.csv
import csv #내장된 csv모듈을 import
import pandas as pd
f= open('train.csv', encoding='utf8') #파일 열기
reader= csv.reader(f) #iterator 타입인 reader객체를 리턴-> for루프를 돌며 한 라인씩 가져올 수 있음// 리턴되는 각 라인은 컬럼들을 나열한 list타입
csv_list=[]
for i in reader:
  csv_list.append(i)
f.close()
train_df= pd.DataFrame(csv_list)
train_df.info()
```

```python
# worked_year 변수지정
train_df['worked_year'] = train_df['DAYS_EMPLOYED']*(-1)
print(train_df['worked_year'].value_counts())
# 취업되지 않은 사람들에 대해 -365 지정
train_df['worked_year'][train_df['worked_year']<0] = -365
print()
print(train_df['worked_year'].value_counts())
# 근무연수를 구하기 위해 worked_year를 365로 나누고 몫만 지정
# 근무연수가 없는 사람들은 -1에 지정
train_df['worked_year'] = train_df['worked_year']//365
print()
print(train_df['worked_year'].value_counts())
```

```python
print(train_df['worked_year'].describe())
#-1: 근무 연수 x, 0:1년 안된 사람 1:1~10년 2: 11년~20년 3: 21년 이상
train_df['worked_year'][(train_df['worked_year']>0) &(train_df['worked_year']<11)] = 1
print(train_df['worked_year'].value_counts())
train_df['worked_year'][(train_df['worked_year']>10)&(train_df['worked_year']<21)] = 2
print(train_df['worked_year'].value_counts())
train_df['worked_year'][(train_df['worked_year']>20)] = 3
print(train_df['worked_year'].value_counts())
```

```python
print(train_df['worked_year'].value_counts())
labels= train_df['worked_year'].value_counts().index.tolist()
ratio=train_df['worked_year'].value_counts().values.tolist()
plt.title('worked_year')
plt.pie(ratio, labels=labels, autopct= '%.1f%%')
plt.show()

#근무 연수 1~10년인 사람이 가장 많이 차지
#10년 이상으로 합칠지는 생각해보기 (3, 4번)
```
